```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Crear datos simulados de serie de tiempo
np.random.seed(42)
data = np.cumsum(np.random.randn(100))

# Ajustar modelo ARIMA
model = ARIMA(data, order=(2,1,2))
model_fit = model.fit()

# Predicción
pred = model_fit.forecast(steps=10)

# Graficar
plt.plot(data, label="Datos")
plt.plot(range(len(data), len(data)+10), pred, label="Predicción", color='red')
plt.legend()
plt.show()
```
